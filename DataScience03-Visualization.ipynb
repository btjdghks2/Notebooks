```python
# mtcats 데이터셋을 이용한 산점도 그래프(Scatter plot)
import pandas as pd
mtcars = pd.read_csv("./data/mtcars.csv") # 헤더가 있고, 구분자가 컴마(,)
mtcars
```

```python
from IPython.display import display
# EDA
print("Size:", mtcars.shape) # 레코드는 32개, 컬럼수 12
# 위쪽 데이터 확인
display(mtcars.head())
display(mtcars.tail())
```

```python
# 기초 통계량 확인
mtcars.describe()
```

```python
# mpg(연비) vs wt(중량) -> 산점도 그려보기
plt.scatter(mtcars['wt'],mtcars['mpg'])
plt.xlabel("weight")
plt.ylabel("Miles per Gallon")
plt.title("weight vs MPG")

plt.show() # 그리기
```

```python
# 상관계수
print("상관계수(wt vs mpg)")
mtcars.corr() # 상관계수 - 모든 수치 변수들 사이의 산관 계수
mtcars['wt'].corr(mtcars['mpg']) # wt 변수와 mpg변수 사이의 상관계수 -> 역상관 관계
```

```python
# 상자 그림(BOx Plot)
#plt.boxplot(mtcars['mpg']) : 단일변수의 상자 그림
plt.boxplot([mtcars['mpg'],mtcars['wt']])
plt.show()

print("mpg 기초 통계량")
display(mtcars['mpg'].describe())
```

```python
# 분위수 확인
quartiles = mtcars.quantile([0.25,0.5,0.75]) # 모든 변수의 분위수
display(quartiles)
```

```python
# mpg의 1, 3사분위수
first_quartile = mtcars['mpg'].quantile(0.25) # 1사 분위수
third_quartile = mtcars['mpg'].quantile(0.75) # 3사 분위수
print("1사분위수 {}. 3사분위수{}".format(first_quartile, third_quartile))
# IQR(inter Quartile Range) : Q3 - Q1
iqr = third_quartile  - first_quartile
print("IQR of mpg:",iqr)
```

```python
# mpg의 상단 극단치경계
# Q3 + IQR * 1.5
border_top = third_quartile + iqr *1.5
print("상단 극단치 경계:",border_top)

border_bottom = third_quartile - iqr *1.5
print("하단 극단치 경계:",border_bottom)
```

```python
# mpg가 극단치를 가진 자동차?
# mpg가 상단 극단치를 초과한 레코드
#mtcars['mpg'] > border_top # boolean 연산
outliers = mtcars[mtcars['mpg'] > border_top]
print("Outliers")
display(outliers)
```
